# Import Libraries

```python
import pandas as pd
import numpy as np
import ast
```

# Load Data

```python
# Read in checkpoint 3
df = pd.read_csv('checkpoint_3.csv')
# df_read_in.info()
df['lat_lon'] = df['lat_lon'].apply(ast.literal_eval).apply(lambda x: (float(x[0]), float(x[1]))) # Change to tuple of floats
```

# Overview of data

```python
df.describe()
```

```python
numeric_columns = df.describe().columns
category_columns = all_columns[~all_columns.isin(numeric_columns)]
category_columns
```

```python
for i in numeric_columns:
    x = pd.DataFrame(df[i].value_counts(normalize=True))
    print(f'The distribution for {i} is {x}')
```
